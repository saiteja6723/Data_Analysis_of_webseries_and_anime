# Naruto Shippuden Episodes Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\megha\\OneDrive\\Documents\\datasets\\naruto.csv",encoding='cp1252')
df.head()
```

```python
df.info()
```

```python
df['Type'].value_counts()
```

```python
df.groupby('Saga' ,as_index = False, sort =False)['Num_episode'].count()
```

```python
season = df['Saga']
for i in range(0,len(season)):
    season[i] = season[i][0:3]
season.value_counts().count()
```

```python
df['Season'] = season
df.head()
```

Checking the number of episodes in each arc

```python
x = df.groupby('Season', as_index=False ,sort=False)['Num_episode'].count()
x
```

```python
plt.figure(figsize=(10,8))
p = sns.countplot(df['Season'])
plt.title("Episodes per season")
```

```python
plt.figure(figsize=(7,5))
plt.plot(x['Season'],x['Num_episode'] , marker = "o")
plt.title("Plot of episodes per season")
```

Count of Canon , Fillers , Mixed-Canon

```python
sns.countplot(df['Type'])
```

```python
rating_mean = df.groupby('Season',sort=False)['Rate'].sum()/df.groupby('Season',sort=False)['Saga'].count()
rating_mean.head()
```

```python
plt.figure(figsize=(7,5))
plt.plot(rating_mean, marker='*',color='y',markerfacecolor='blue')
plt.title("Mean Rating of each season")
plt.xlabel("Average Rating per season")
```

```python
sns.scatterplot(df['Num_episode'],df['Type'] , hue =df['Saga'])
plt.show()
```

```python
sns.catplot(data=df, x="Rate", y="Season", hue="Type")
plt.title("Rate by Seasons")
```

```python
maximum = max(df.Rate) 
minimum = min(df.Rate)
maximum,minimum
```

```python
df.query(" Rate == [9.5,4.2]")
```

Highest and lowest Rated Episode in each Season

```python
max_season = df.groupby("Season" , sort = False)['Rate'].max()
min_season = df.groupby("Season",sort =False)['Rate'].min()
```

```python
plt.figure(figsize=(7,5))
plt.plot(max_season,marker='o')
plt.plot(min_season,marker='>')
plt.title("Highest and Lowest Rating of each season")
plt.xlabel("Season")
plt.ylabel("Rating")
plt.legend(["Max","Min"])
```

Episodes Released per year

```python
year_episode = df.groupby('Year_launch',as_index=False,sort = False)['Num_episode'].count()
seasons = [x for x in range(2007,2018)]
```
